```python
import time

# Inicializa a lista para armazenar todas as perguntas geradas
questions = []

# Supondo que `documents` seja uma lista de documentos e você está interessado em processar um subset específico
start = 7000
end = 9988
documents_subset = documents[start:end]

# Variáveis para controlar a taxa de requisição
requests_per_minute = 100
delay_between_requests = 10  # Ajustado para 10 segundos conforme a mensagem de erro

# Loop para processar cada documento em `documents_subset`
for i, document in enumerate(documents_subset):
    try:
        if i > 0 and i % requests_per_minute == 0:
            print(f"Pausing for {delay_between_requests} seconds to limit requests to {requests_per_minute} per minute.")
            time.sleep(delay_between_requests)  # Pausa o script por 10 segundos
        
        # Gera perguntas para o documento atual
        train_dataset_generator = RagDatasetGenerator.from_documents(
            [document],  # Passa o documento atual no loop
            question_gen_query=QUESTION_GEN_PROMPT,
            llm=llm,
            show_progress=True,
            num_questions_per_chunk=1,
        )
        train_questions = train_dataset_generator.generate_questions_from_nodes()
        
        # Adiciona as perguntas geradas à lista `questions`
        questions.extend([e.query for e in train_questions.examples])

        print(f"Processed document {i+start} (overall document index)")
    except Exception as e:
        print(f"Exception occurred: {e}")
        print("Waiting for 10 seconds before retrying...")
        time.sleep(10)

# Ao final do loop, `questions` contém todas as perguntas geradas para os documentos processados
```

```python
# Transformação e Exportação - dataset labeled

import csv

# Preparando para escrever no arquivo CSV com o novo delimitador e coluna adicional
with open('questions_labeled_500:1500.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=';')  # Usando ';' como delimitador
    # Escrevendo os cabeçalhos das colunas
    csvwriter.writerow(['prompt', 'response', 'label'])

    for i in range(0, len(questions), 8):  # Iterar através da lista considerando 8 itens por conjunto
        # Verifica se o índice do último elemento esperado está dentro da lista
        if i + 7 < len(questions):
            # Removendo o prefixo "Pergunta: " e limpando espaço extra
            prompt = questions[i].replace('Pergunta: ', '').strip()

            # A segunda posição é a resposta correta, removendo o prefixo e limpando espaço extra
            correct_response = questions[i+1].replace('Resposta correta: ', '').strip()
            
            # Escrevendo a pergunta e a resposta correta com label 1
            csvwriter.writerow([prompt, correct_response, 1])

            # As posições 4 a 8 são as respostas 'hard negative'
            for j in range(3, 8):  # Iterando pelas respostas 'hard negative'
                hard_negative = questions[i+j].strip()
                # Escrevendo a pergunta e cada resposta 'hard negative' com label -1
                csvwriter.writerow([prompt, hard_negative, -1])

            
print("Arquivo CSV criado com sucesso.")
```

```python
import time
from llama_index.core.evaluation import DatasetGenerator
from llama_index.core.llama_dataset.generator import RagDatasetGenerator
import pandas as pd

# Inicializa a lista para acumular os resultados
dataset = []

start = 8000
end = 9000   
documents_subset = documents[start:end]

# Variáveis para controlar a taxa de requisição
requests_per_minute = 100
delay_between_requests = 10

# Loop para processar cada documento em `documents_subset`
for i, document in enumerate(documents_subset):
    try:
        if i > 0 and i % requests_per_minute == 0:
            print(f"Pausing for {delay_between_requests} seconds to limit requests to {requests_per_minute} per minute.")
            time.sleep(delay_between_requests)
        
        rag_dataset_generator = RagDatasetGenerator.from_documents(
            [document],
            question_gen_query=QUESTION_GEN_PROMPT,
            llm=llm,
            show_progress=True,
            num_questions_per_chunk=1,
        )
        rag_dataset = rag_dataset_generator.generate_dataset_from_nodes()
        
        # Extrai e adiciona os dados de cada exemplo ao lista_dataset
        for example in rag_dataset.examples:
            data = {
                'query': example.query,
                'reference_contexts': example.reference_contexts,
                'reference_answer': example.reference_answer,
                'reference_answer_by': example.reference_answer_by.model_name if example.reference_answer_by else None,
                'query_by': example.query_by.model_name if example.query_by else None,
            }
            dataset.append(data)

        print(f"Processed document {i+start} (overall document index)")
        
    except Exception as e:
        print(f"Exception occurred: {e}")
        print("Waiting for 10 seconds before retrying...")
        time.sleep(10)
```
