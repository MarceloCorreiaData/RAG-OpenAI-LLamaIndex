## Implementação do Hybrid Retriever (Embedding/Vector Retriever + BM25 algorithm Retreiver)

```python
from llama_index.retrievers import BaseRetriever
```

```python
class HybridRetriever(BaseRetriever):
    def __init__(self, vector_retriever, bm25_retriever):
        self.vector_retriever = vector_retriever
        self.bm25_retriever = bm25_retriever
        super().__init__()

    def _retrieve(self, query, **kwargs):
        bm25_nodes = self.bm25_retriever.retrieve(query, **kwargs)
        vector_nodes = self.vector_retriever.retrieve(query, **kwargs)

        # combine the two lists of nodes
        all_nodes = []
        node_ids = set()
        for n in bm25_nodes + vector_nodes:
            if n.node.node_id not in node_ids:
                all_nodes.append(n)
                node_ids.add(n.node.node_id)
        return all_nodes
```

```python
# Instanciando hybrid_retriever a partir das 2 instâncias anteriores (vector_retriever, bm25_retriever)
hybrid_retriever = HybridRetriever(vector_retriever, bm25_retriever)
```

```python
import re

def limpar_texto(source_string):
    # Remove o padrão 'Source (Doc id: ...): ' do início
    clean_text = re.sub(r'^Source \(Doc id: [a-z0-9-]+\): ', '', source_string)
    
    # Remove os colchetes e as quebras de linha
    clean_text = re.sub(r'\[\n|\n\]', '', clean_text)  # Remove os colchetes com quebras de linha
    clean_text = clean_text.replace('\n', '')  # Remove todas as quebras de linha restantes
    
    # Remove o padrão ': []' que sobrou ao final das strings
    clean_text = re.sub(r': \[\]$', '', clean_text)
    
    # Opcional: remover espaços extras resultantes da remoção
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    
    return clean_text

# Aplicar a função limpar_texto a cada item em results
source_texts = [limpar_texto(result) for result in results]
```

```python
from transformers import AutoTokenizer, AutoModel

model_name = 'BAAI/bge-large-en-v1.5'  
model = AutoModel.from_pretrained(model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Gerar embedding para o texto da query diretamente

# Função para pooling médio
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0]  # Primeiro elemento contém os embeddings dos tokens
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

# Função para gerar os vetores de embeddings
def gerar_embedding(texto):
    inputs = tokenizer(texto, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        model_output = model(**inputs)
        # Aplicar pooling médio e normalizar
        sentence_embedding = mean_pooling(model_output, inputs['attention_mask'])
        return sentence_embedding[0].numpy()

def gerar_embeddings_com_hierarquia(json_data, path=''):
    embeddings_hierarquicos = {}
    for chave, valor in json_data.items():
        novo_caminho = f"{path}/{chave}" if path else chave
        if isinstance(valor, dict):
            embeddings_hierarquicos.update(gerar_embeddings_com_hierarquia(valor, novo_caminho))
        elif isinstance(valor, list):
            for texto in valor:
                embedding = gerar_embedding(texto)
                embeddings_hierarquicos[novo_caminho] = embedding
        elif isinstance(valor, str):
            embedding = gerar_embedding(valor)
            embeddings_hierarquicos[novo_caminho] = embedding
    return embeddings_hierarquicos    
```

```python
def processar_jsons_lista(json_lista):
    embeddings_hierarquicos_total = {}  # Este dicionário vai conter todos os embeddings
    for i, json_element in enumerate(json_lista):
        # Gera um novo caminho base para cada elemento da lista, usando o índice
        novo_caminho_base = str(i)
        # Chama a função original para cada dicionário JSON na lista
        embeddings_hierarquicos = gerar_embeddings_com_hierarquia(json_element, novo_caminho_base)
        embeddings_hierarquicos_total.update(embeddings_hierarquicos)  # Atualiza o dicionário total com os novos embeddings
    return embeddings_hierarquicos_total

# Para processar o JSON que é uma lista de JSONs combinados
if isinstance(json_data, list):
    embeddings_hierarquicos = processar_jsons_lista(json_data)
```

## Riemannian Distribution approximation and projection of Embeddings (2D)

```python
# Redução de Dimensionalidade - Preparação dos dados 2D
umap_2d = umap.UMAP(n_neighbors=15, n_components=2, metric='euclidean', random_state=42).fit(embeddings_combined)
```

```python
# Função de Projeção - p/ projetar os embeddings do dataset, da query e das sources em 2 dimensões

def project_embeddings(embeddings, umap_transform):
    # Inicializa um array numpy vazio com o comprimento dos embeddings e 2 dimensões
    umap_embeddings = np.empty((len(embeddings), 2))
    # Projeta cada embedding individualmente
    for i, embedding in enumerate(tqdm(embeddings)):
        umap_embeddings[i] = umap_transform.transform([embedding])[0] 
    return umap_embeddings
```
